```python
x = [1, 2, 3]
y = [1, 2, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("X & Y Ejemplo ")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Este es y", "este es z"])
plt.show()
```
